## Дерева класифікації та регресії. Особливості, обмеження та переваги

**Дерева класифікації** та регресії називають також деревами рішення. Ці моделі поступово ділять початковий простір спостережень на декілька підпросторів. Тобто на кожній ітерації деяка характеристика вхідного спостереження порівнюється з певним критерієм і простір ділиться далі в залежності від отриманого результату. Після завершення такого поділу ми отримуємо відгук, усереднений по отриманих "областях":
$$f(x) = E(y|x) = \Sigma_{m=1}^Mw_mI(x \in R_m),$$
де $R_m$ - m-та область, $w_m$ - середній відгук в ній.

**Переваги:**
1. Їх легко інтерпретувати і пояснювати, чому ж було прийнято саме таке рішення
2. Вони легко працюють одначасно і з дискретними, і з неперервними значеннями
3. Відносно не чутливі до викидів
4. Можуть бути змодифіковані таким чином, щоб справлятися із відсутніми даними

**Недоліки:**
1. Їх прогноз не такий точний порівняно з іншими моделями
2. Є нестабільними, тобто деяка зміна у вхідних даних може сильно вплинути на структуру дерева.
Щоб побороти ці недоліки, наприклад, застосовують метод агрегації даних, будують випадкові ліси, що складається з декількох дерев і т.п. 



##Матриця Грама

Обчислити матрицю Грама для ядерної функції 

k(x y) = y * cos(x) + x^3 / 4 

на множині X, Y = {1,4,2,5,3,7}. 

Аргумент функції sin(x) задано в радіанах.

```
import numpy as np
def k(x,y):
  return y*np.cos(x)+(x**3)/4
  

def gram(X, Y):
  n = len(X)
  gram = [[k(x, y)*k(x, y) for x in X] for y in Y]
  return gram
```

```
X=[1,4,2,5,3,7]
Y=X
gram(X, Y)
```
