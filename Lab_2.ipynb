```python
import numpy as np
import pandas as pd
import random
import os
import cv2
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
import glob
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
```

```python
IMAGE_COUNT = 250
NICCKY_NUMBER = 8
FULL_IMAGES_COUNT = len(os.listdir('mirflickr'))
```

## 1)	Сформувати тестову вибірку зображень з вихідного пакету;

```python
images = []
while len(images) != IMAGE_COUNT:
    randIndex = random.randint(NICCKY_NUMBER, FULL_IMAGES_COUNT)
    image_name = 'im{index}.jpg'.format(index = randIndex)
    image = cv2.imread(glob.glob('./mirflickr/{}'.format(image_name))[0])[:, :, 1]
    images.append(image)
    
images = np.asarray(images)
```

## 2)	Провести декомпозицію каналу зеленого кольору тестових зображень з використанням методу головних компонент (PCA)

```python
def pcaProcessing(image, components = None, invTransform = True):
    pca = PCA(n_components=components, svd_solver='full')
    res = pca.fit_transform(image)
    if invTransform:
        return pca.inverse_transform(res)
    return pca, res
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(images[10])

plt.subplot(1,2,2)
plt.title('Reconstructed')
plt.imshow(pcaProcessing(images[10]))
```

## 3)	Для каналу зеленого кольору тестових зображень обчислити наступні характеристики розподілу значень яскравості пікселів:

####	Впорядкувати отримані компоненти вихідного зображення в порядку зменшення значень сингулярних чисел (від найбільшого s_max до найменшого s_min значення);

```python
pca, compressed = pcaProcessing(images[10], invTransform= False)
pca.singular_values_
```

#### Провести реконструкцію зображення при використанні лише частки (α%) компонентів розкладу, що характеризуються відмінними від нуля сингулярними числами (s_i≠0). Розглянути випадок, коли α змінюється від α_min=10% до α_max=100% з кроком ∆_α=10%.

#### Порівняти вихідне та реконструйоване зображення за показником середньоквадратичного відхилення (MSE). 

```python
rowIndex = 1
for image in images[:2]:
    plt.figure(figsize=(32, 18))
    mseValues = []
    for k in range(1, 11):
        res = pcaProcessing(image, components = int(min(np.shape(image)[0], np.shape(image)[1]) * k * 0.1))
        mse = mean_squared_error(image, res)
        mseValues.append(mse)
        
        plt.subplot(rowIndex,11,k+1)
        plt.title('image, mse: {} %'.format(k*10))
        plt.imshow(image)
    print('MSE: {}'.format(mseValues))
    plt.subplot(rowIndex,11,1)
    plt.title('MSE')
    plt.plot(mseValues)
    rowIndex+=1
    
    
    
    
```

#### Побудувати графіки залежності (MSE) ̅(α), де (MSE) ̅- значення середньо-квадратичного відхилення між вихідним та реконструйованим зображеннями, усереднені по тестовому пакету;

```python
def calcMeanMSE(images):
    resMSE = np.zeros(10,)
    for image in images:
        mseValues = []
        for k in range(1, 11):
            res = pcaProcessing(image, components = int(min(np.shape(image)[0], np.shape(image)[1]) * k * 0.1))
            mse = mean_squared_error(image, res)
            mseValues.append(mse)
            
        np.add(resMSE, np.array(mseValues), out = resMSE)
    return resMSE / len(images)
```

```python
meanMSE = calcMeanMSE(images)
plt.figure(figsize=(8, 6))
plt.title('Mean MSE')
plt.plot(meanMSE, label='MSE $\\alpha$')
plt.legend()
plt.show();
```

## Провести моделювання каналу зеленого кольору тестових зображень з використанням марківських ланцюгів першого порядку M_1:
#### Отримати стохастичну матрицю для каналу зеленого кольору при обробці пікселів (згідно номеру студента в списку групи, за модулем кількості варіантів):
#### По горизонталі, зліва направо – M_1^→ (I_(x,y),I_(x+1,y) );
#### По горизонталі, справа наліво – M_1^← (I_(x,y),I_(x-1,y) );
#### По вертикалі, зверху вниз – M_1^↓ (I_(x,y),I_(x,y+1) );
#### По вертикалі, знизу вгору – M_1^↑ (I_(x,y),I_(x,y-1) );
#### По головній діагоналі – M_1^↘ (I_(x,y),I_(x+1,y+1) );
#### По головній діагоналі – M_1^↖ (I_(x,y),I_(x-1,y-1) );
#### По побічній діагоналі – M_1^↙ (I_(x,y),I_(x-1,y+1) );
#### По побічній діагоналі – M_1^↗ (I_(x,y),I_(x+1,y-1) );


```python
def normalizeMatrix(matrix):
    rowSum = matrix.sum(axis = 1)
    res = matrix / rowSum[:, np.newaxis]
    return res
```

```python
def imageIterL_R_and_R_L(image, matrix, rightToLeft = False):
    height, width = np.shape(image)
    for h in range(height):
        for w in range(width - 1):
            if not rightToLeft:
                c1 = image[h][w]
                c2 = image[h][w + 1]
                matrix[c2][c1] += 1
            else:
                c1 = image[h][width - w - 2]
                c2 = image[h][width - w - 1]
                matrix[c1][c2] += 1
    return matrix
```

```python
def imageIterT_D_and_D_T(image, matrix,downToTop = False):
    height, width = np.shape(image)
    for h in range(height - 1):
        for w in range(width):
            if not downToTop:
                c1 = image[h][w]
                c2 = image[h + 1][w]
                matrix[c2][c1] += 1
            else:
                c1 = image[height - h - 2][w]
                c2 = image[height - h - 1][w]
                matrix[c1][c2] += 1
    return matrix
```

```python
def isRegular(matrix, k):
    m = np.linalg.matrix_power(matrix, k)
    return not ((m <= 0).sum() != 0)
```

```python
def isRecurence(matrix, k):
    m = np.linalg.matrix_power(matrix, k)
    for i in range(np.shape(m)[0]):
        if m[~i, i] == 0:
            return False
        return True
```

#### В протоколі роботи графічно показати вид марківського ланцюга для діапазону яскравості пікселів I_(x,y)∈[i;i×10], де i- номеру студента в списку групи;
#### Для отриманих марківських ланцюгів перевірити виконання властивості регулярності та рекурентності після проходження 5 ітерацій роботи.

```python

####
stochMatrix = np.zeros((256, 256))  
for image in images:
    imageIterL_R_and_R_L(image, stochMatrix)
stochMatrix = normalizeMatrix(stochMatrix)
plt.figure(figsize=(16, 9))
print('Matrix is regular: {}'.format(isRegular(stochMatrix, k = 5)))
print('Marix is recurrent: {}'.format(isRecurence(stochMatrix, k = 5)))
plt.title("Left to Right")
sns.heatmap(stochMatrix[NICCKY_NUMBER:NICCKY_NUMBER * 10, NICCKY_NUMBER:NICCKY_NUMBER * 10], square=True, cmap='Blues')
plt.show();

####
stochMatrix = np.zeros((256, 256))  
for image in images:
    imageIterL_R_and_R_L(image, stochMatrix,rightToLeft = True)
stochMatrix = normalizeMatrix(stochMatrix)
plt.figure(figsize=(16, 9))
print('Matrix is regular: {}'.format(isRegular(stochMatrix, k = 5)))
print('Marix is recurrent: {}'.format(isRecurence(stochMatrix, k = 5)))
plt.title("Right to Left")
sns.heatmap(stochMatrix[NICCKY_NUMBER:NICCKY_NUMBER * 10, NICCKY_NUMBER:NICCKY_NUMBER * 10], square=True, cmap='Blues')
plt.show();

####
stochMatrix = np.zeros((256, 256))  
for image in images:
    imageIterT_D_and_D_T(image, stochMatrix)
stochMatrix = normalizeMatrix(stochMatrix)
plt.figure(figsize=(16, 9))
print('Matrix is regular: {}'.format(isRegular(stochMatrix, k = 5)))
print('Marix is recurrent: {}'.format(isRecurence(stochMatrix, k = 5)))
plt.title("To to down")
sns.heatmap(stochMatrix[NICCKY_NUMBER:NICCKY_NUMBER * 10, NICCKY_NUMBER:NICCKY_NUMBER * 10], square=True, cmap='Blues')
plt.show();

####
stochMatrix = np.zeros((256, 256))  
for image in images:
    imageIterT_D_and_D_T(image, stochMatrix,downToTop=True)
stochMatrix = normalizeMatrix(stochMatrix)
plt.figure(figsize=(16, 9))
print('Matrix is regular: {}'.format(isRegular(stochMatrix, k = 5)))
print('Marix is recurrent: {}'.format(isRecurence(stochMatrix, k = 5)))
plt.title("Down to top")
sns.heatmap(stochMatrix[NICCKY_NUMBER:NICCKY_NUMBER * 10, NICCKY_NUMBER:NICCKY_NUMBER * 10], square=True, cmap='Blues')
plt.show();
```

```python
def firstDiagT_D_and_D_T(image, matrix,downTop = False):
    height, width = np.shape(image)
    for h in range(height - 1):
        for w in range(width - 1):
            if not downTop:
                c1 = image[h][w]
                c2 = image[h + 1][w + 1]
                matrix[c2][c1] += 1
            else:
                c1 = image[height - h - 2][width - w - 2]
                c2 = image[height - h - 1][width - w - 1]
                matrix[c1][c2] += 1
    return matrix
```

```python
def secondDiagT_D_and_D_T(image, matrix,downTop = False):
    height, width = np.shape(image)
    for h in range(height - 1):
        for w in range(width - 1):
            if not downTop:
                c1 = image[h][width - w - 1]
                c2 = image[h + 1][width - w - 2]
                matrix[c2][c1] += 1
            else:
                c1 = image[height - h - 1][w]
                c2 = image[height - h - 2][w + 1]
                matrix[c2][c1] += 1
    return matrix
```

```python
####
stochMatrix = np.zeros((256, 256))  
for image in images:
    firstDiagT_D_and_D_T(image, stochMatrix)
stochMatrix = normalizeMatrix(stochMatrix)
plt.figure(figsize=(16, 9))
print('Matrix is regular: {}'.format(isRegular(stochMatrix, k = 5)))
print('Marix is recurrent: {}'.format(isRecurence(stochMatrix, k = 5)))
plt.title("Diag top to down")
sns.heatmap(stochMatrix[NICCKY_NUMBER:NICCKY_NUMBER * 10, NICCKY_NUMBER:NICCKY_NUMBER * 10], square=True, cmap='Blues')
plt.show();

####
stochMatrix = np.zeros((256, 256))  
for image in images:
    firstDiagT_D_and_D_T(image, stochMatrix,downTop = True)
stochMatrix = normalizeMatrix(stochMatrix)
plt.figure(figsize=(16, 9))
print('Matrix is regular: {}'.format(isRegular(stochMatrix, k = 5)))
print('Marix is recurrent: {}'.format(isRecurence(stochMatrix, k = 5)))
plt.title("Diag down to top")
sns.heatmap(stochMatrix[NICCKY_NUMBER:NICCKY_NUMBER * 10, NICCKY_NUMBER:NICCKY_NUMBER * 10], square=True, cmap='Blues')
plt.show();

####
stochMatrix = np.zeros((256, 256))  
for image in images:
    secondDiagT_D_and_D_T(image, stochMatrix)
stochMatrix = normalizeMatrix(stochMatrix)
plt.figure(figsize=(16, 9))
print('Matrix is regular: {}'.format(isRegular(stochMatrix, k = 5)))
print('Marix is recurrent: {}'.format(isRecurence(stochMatrix, k = 5)))
plt.title("Second diag top to down")
sns.heatmap(stochMatrix[NICCKY_NUMBER:NICCKY_NUMBER * 10, NICCKY_NUMBER:NICCKY_NUMBER * 10], square=True, cmap='Blues')
plt.show();

####
stochMatrix = np.zeros((256, 256))  
for image in images:
    secondDiagT_D_and_D_T(image, stochMatrix,downTop=True)
stochMatrix = normalizeMatrix(stochMatrix)
plt.figure(figsize=(16, 9))
print('Matrix is regular: {}'.format(isRegular(stochMatrix, k = 5)))
print('Marix is recurrent: {}'.format(isRecurence(stochMatrix, k = 5)))
plt.title("Second diag down to top")
sns.heatmap(stochMatrix[NICCKY_NUMBER:NICCKY_NUMBER * 10, NICCKY_NUMBER:NICCKY_NUMBER * 10], square=True, cmap='Blues')
plt.show();
```
